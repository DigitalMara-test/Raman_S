# üß¨ Price Optimization with Evolutionary Algorithm

## üéØ Objective

To maximize **profit** defined by:

$
\text{Profit}(p) = (p - c) \cdot q(p) \cdot (1 - \text{churn}(p))
$

Where:
- `p` = `our_price`
- `c` = `production_cost`
- `q(p)` = Predicted quantity (XGBoost regression model)
- `churn(p)` = Predicted churn probability (XGBoost classification model)

## üõ†Ô∏è Methodology

An **Evolutionary Algorithm (EA)** was implemented for non-linear, constrained, black-box optimization. It operates as a global optimizer suited for non-differentiable functions where multiple outputs are evaluated via trained ML models.

### Key Components:
- **Population Initialization**: 90 individuals per generation
- **Selection**: Tournament selection
- **Crossover**: Simulated Binary Crossover (SBX)
